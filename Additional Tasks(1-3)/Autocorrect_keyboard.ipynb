# Autocorrect Keyboard System

## Objective:
Develop an autocorrect system that identifies misspelled words and automatically corrects them to improve text input accuracy.

## Tools Used:
- Python
- pyspellchecker library

## Outcome:
The system detects incorrect words and provides automatic corrections in real-time.

```python
from spellchecker import SpellChecker

spell = SpellChecker()
```

```python
text = "I lov machne lernng"

words = text.split()

misspelled = spell.unknown(words)

print("Misspelled words:", misspelled)
```

```python
for word in misspelled:
    print("Wrong word:", word)
    print("Suggested correction:", spell.correction(word))
    print()
```

```python
def autocorrect_sentence(sentence):
    words = sentence.split()
    corrected_words = []
    
    for word in words:
        if word in spell:
            corrected_words.append(word)
        else:
            corrected_words.append(spell.correction(word))
    
    return " ".join(corrected_words)
```

```python
sentence = input("Type your sentence: ")
print("Corrected Sentence:", autocorrect_sentence(sentence))
```
